Keşifçi veri analizi:

veri bilimcilerin ellerindeki verinin genel özelliklerini resmetmek için kullandıkları yöntemler bütününe verilen addır. 
Sıklıkla görselleştirme metotları kullanılıyor olsa da istatistik fonksiyonlar, pivot tablolar da keşifçi veri analizi yapılırken kullanılabilir. 
Bu analizi yapmaktaki amaç temelde değişkenleri ve değişkenlerin birbirleriyle ilişkilerini gözlemlemektir.

Türleri:

1 - Tek değişken analizi: Tek bir sütunla uğraşıldığı için çıkarım yapılacak veri sayısı en azdır. 
Analize yalnızca tek bir sütunun dahil olması sebeb veya regresyon analizinden çok sütunun dağılımını ve genel özelliklerini görmeyi gerektirir.

2 - İki değişken analizi: İki değişken arasındaki ilişkinin sorgulandığı ve muhtmemel bir sebep sonuç bağlantısına gidilmek istenen analizdir.

3 - Çok değişken analizi: Üç ve daha fazla sayıda değişkenin analiz edildiği yöntemdir.


Değişken Analizleri Analiz Yöntemine Göre de İkiye Ayrılır:

1- Grafikli Analiz: Verideki eğilimler ve örüntüleri görmek için tablolar ve grafikler kullanılan yöntemdir. Örn: matplotlib ile yapılan grafikler

2 - Grafiksiz Analiz: Görselleştirme olmaksızın istatistiki araçlar kullanılarak yapılan analizdir. Örn: mod, medyan, çarpıklık (skewness)

Keşifçi Veri Analizi Örnekleri:

-> Gruplama (clustering) ve boyut düşürme (dimensionality reduction) yöntemleri ile görselleştirmeler.
-> Tek değişkenli analiz içindeki sütunların tek tek özet istatistiklerle de desteklenerek görselleştirilmesi.
-> İki değişkenli analiz ile değişkenlerin birbirleri arasındaki ilişkinin ve korelasyonun incelenmesi.
