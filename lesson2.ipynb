1- Python'da Veri Görselleştirme Kütüphaneleri

a- Matplotlib (Veri Görselleştirme)

```python
import matplotlib.pyplot as plt  # ilk olarak matplotlib kütüphanesini çağırıyoruz.
```

Örnek:

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11] # Değişkenleri tanımladık.
```

Çizgi Grafiği

```python
plt.figure(figsize=(8,4)) # Grafiğin boyutunu ayarlar
```

```python
plt.plot(x, y, marker='o',color='blue',linestyle='-') # Çizgi grafiğini oluşturur
```

```python
plt.title('Örnek çizgi grafiği') # Grafiğe başlık ekler
```

```python
plt.xlabel('X Ekseni') # x eksen etiketi
```

```python
plt.ylabel('Y Ekseni') # Y eksen etiketi
```

```python
plt.grid(True) # Izgara çizgisini ekler
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, marker='o',color='blue',linestyle='-')
plt.title('Örnek çizgi grafiği')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.grid(True)
plt.show() # Grafiği gösterir
```

Sütun Grafiği

```python
plt.figure(figsize=(8,4))
```

```python
plt.bar(x,y,color='green')
```

```python
plt.title('Örnek sütun grafiği')
```

```python
plt.xlabel('X Ekseni')
```

```python
plt.ylabel('Y Ekseni')
```

```python
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8, 4))  # Grafiğin boyutunu ayarlar
plt.bar(x, y, color='green')  # Sütun grafiğini oluşturur
plt.title('Örnek Sütun Grafiği')  # Grafiğe başlık ekler
plt.xlabel('X Ekseni')  # X ekseni etiketi
plt.ylabel('Y Ekseni')  # Y ekseni etiketi
plt.grid(axis='y')  # Y ekseni boyunca izgara çizgisini ekler
plt.show()  # Grafiği gösterir
```

1.	Çizgi Grafiği:
●	plt.plot() fonksiyonu kullanılarak çizgi grafiği oluşturulur.
●	marker parametresiyle noktalar belirtilir, color parametresiyle renk, ve linestyle parametresiyle çizgi stili belirtilir.
●	plt.title(), plt.xlabel(), plt.ylabel() ve plt.grid() fonksiyonlarıyla grafiğe başlık, eksen etiketleri ve izgara çizgisi eklenir.
2.	Sütun Grafiği:
●	plt.bar() fonksiyonu kullanılarak sütun grafiği oluşturulur.
●	color parametresiyle sütunların rengi belirtilir.
●	Diğer aşamalar çizgi grafiği ile benzer şekilde gerçekleştirilir.
