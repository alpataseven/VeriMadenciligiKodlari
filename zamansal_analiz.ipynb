<h1>Zamansal Analiz</h1>

Zaman içindeki değişimleri incelemek için kullanılan bir analiz türüdür. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Tarih': pd.date_range(start='2024-01-01', periods=100),
    'Kapanış Fiyatı': [100, 105, 110, 108, 112, 115, 120, 118, 122, 125,
                       130, 128, 132, 135, 138, 140, 142, 145, 148, 150,
                       155, 160, 158, 162, 165, 168, 170, 172, 175, 180,
                       178, 182, 185, 188, 190, 192, 195, 198, 200, 202,
                       205, 208, 210, 212, 215, 218, 220, 222, 225, 228,
                       230, 232, 235, 238, 240, 242, 245, 248, 250, 252,
                       255, 258, 260, 262, 265, 268, 270, 272, 275, 278,
                       280, 282, 285, 288, 290, 292, 295, 298, 300, 302,
                       305, 308, 310, 312, 315, 318, 320, 322, 325, 328,
                       330, 332, 335, 338, 340, 342, 345, 348, 350, 352]
}
```

```python
# Verileri DataFrame'e dönüştürelim
df = pd.DataFrame(data)
df.set_index('Tarih', inplace=True)  # Tarih sütununu indeks olarak ayarlayalım
```

```python
# Hisse senedinin kapanış fiyatlarını görselleştirelim
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Kapanış Fiyatı'], marker='o', linestyle='-')
plt.title('Hisse Senedi Kapanış Fiyatları (2024)')
plt.xlabel('Tarih')
plt.ylabel('Kapanış Fiyatı ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
