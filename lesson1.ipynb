```python
#Matplotlib örnekleri: 

# 1. Çizgi Grafik Oluşturma:

#Önce matpoltlib kütüphanesini çağırıyoruz

import matplotlib.pyplot as plt
```

```python
# Veri Oluşturma

x = [1,2,3,4,5]
y = [3,6,8,9,7]
```

```python
# Çizgi grafiği oluşturma

plt.plot(x,y)

# Grafiği gösterme

plt.show()
```

```python
# 3. Histogram Oluşturma:

import matplotlib.pyplot as plt3
import numpy as np
```

```python
# Rastgele veri oluşturma

data = np.random.normal(0,1,1000)

```

0: Dağılımın ortalaması (mean)
1: Dağılımın standart sapması (standard deviation)
1000: Üretilecek veri sayısı

```python
# Histogram oluşturma

plt.hist(data, bins=30, edgecolor='yellow', color='red')
```
